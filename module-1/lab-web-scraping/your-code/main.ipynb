####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url7 = 'https://www.emsc-csem.org/Earthquake/'
````

````python
import numpy as np
import pandas as pd

res7 = requests.get(url7)
soup7 = BeautifulSoup(res7.text, 'html.parser')
#print(soup7.prettify())

dates = soup7.select("tbody tr td.tabev6 b a")
#separamos date de time como nos dice el enunciado
date=[]
time=[]

for x in dates:
    date.append(x.text[0:10])
    time.append(x.text[-10:])
latitudes = soup7.select("tbody tr td.tabev1")
#separamos latitud de longitud como nos dice el enunciado
latitude=[]
longitude=[]
for x in latitudes:
    latitude.append(x.text[:-1])
longitudes = soup7.select("tbody tr td.tabev1")
longitude = Latitude[1::2]
latitude = Latitude[0::2]

regions = soup7.select("tbody tr td.tb_region")
region=[]
for x in regions:
    region.append(x.text[1:])

df = pd.DataFrame(list(zip(date,time,longitude,latitude,region)),
              columns=['Dates','Time','Latitude','Longitude','Region'])
display(df)
````
